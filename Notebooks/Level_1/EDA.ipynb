```python
from sklearn.feature_selection import mutual_info_classif
import numpy as np

# Select only numeric columns
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate mutual information scores
mi_scores = mutual_info_classif(df[numeric_cols].fillna(0), df['churn'])

# Create a dataframe with the scores
mi_df = pd.DataFrame({'Feature': numeric_cols, 'MI Score': mi_scores})
mi_df = mi_df.sort_values('MI Score', ascending=False)

# Plot the scores
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(mi_df['Feature'], mi_df['MI Score'])
plt.xticks(rotation=45, ha='right')
plt.title('Mutual Information Scores for Numeric Features vs Churn')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import mutual_info_classif
import numpy as np

import matplotlib.pyplot as plt

# Select only categorical columns (excluding 'churn' which is our target)
categorical_cols = df.select_dtypes(include=['object']).columns.drop('churn')

# Initialize dictionary to store MI scores
mi_scores_cat = {}

# Calculate MI score for each categorical variable
le = LabelEncoder()
for col in categorical_cols:
    # Label encode the categorical column
    encoded_col = le.fit_transform(df[col].fillna('missing'))
    # Calculate MI score
    mi_score = mutual_info_classif(encoded_col.reshape(-1, 1), df['churn'], discrete_features=True)[0]
    mi_scores_cat[col] = mi_score

# Create dataframe of scores and sort
mi_cat_df = pd.DataFrame({'Feature': mi_scores_cat.keys(), 'MI Score': mi_scores_cat.values()})
mi_cat_df = mi_cat_df.sort_values('MI Score', ascending=False)

# Plot
plt.figure(figsize=(12,6))
plt.bar(mi_cat_df['Feature'], mi_cat_df['MI Score'])
plt.xticks(rotation=45, ha='right')
plt.title('Mutual Information Scores for Categorical Features vs Churn')
plt.tight_layout()
plt.show()

# Display the scores
print("\nMutual Information Scores for Categorical Features:")
print(mi_cat_df)
```

```python
# Create temporary numeric mapping for satisfaction scores
satisfaction_num_map = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}

# Create temporary numeric mapping for complain
complain_num_map = {'Yes': 1, 'No': 0}

# Create net satisfaction column
df['net_satisfaction'] = df['satisfactionscore'].map(satisfaction_num_map) - df['complain'].map(complain_num_map)
```

```python
# Create mapping for net satisfaction scores
net_satisfaction_mapping = {
    1: 'One',
    2: 'Two',
    3: 'Three',
    4: 'Four',
    5: 'Five',
    0: 'Zero',
    -1: 'Negative One'
}

# Apply mapping to net_satisfaction column
df['net_satisfaction'] = df['net_satisfaction'].map(net_satisfaction_mapping)
```

```python
eda_helper.cat_bivar_plots(df,'net_satisfaction','churn')
```

```python
# Calculate MI score for net_satisfaction
encoded_col = le.fit_transform(df['net_satisfaction'].fillna('missing'))
mi_score = mutual_info_classif(encoded_col.reshape(-1, 1), df['churn'], discrete_features=True)[0]

# Print the result
print(f"Mutual Information Score for net_satisfaction vs churn: {mi_score:.6f}")

# Plot for comparison
plt.figure(figsize=(6,4))
plt.bar(['net_satisfaction'], [mi_score])
plt.title('Mutual Information Score: net_satisfaction vs Churn')
plt.tight_layout()
plt.show()
```
